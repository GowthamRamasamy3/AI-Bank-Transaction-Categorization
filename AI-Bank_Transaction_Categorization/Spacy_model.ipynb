```python
import pandas as pd
import spacy
from google.colab import files
import matplotlib.pyplot as plt
import seaborn as sns

# Load spaCy model
nlp = spacy.load('en_core_web_sm')

# Load categories and keywords
categories_df = pd.read_csv('/content/categories_and_keywords (1).csv')

# Convert keywords to lowercase for case-insensitive comparison
categories_df['Keywords'] = categories_df['Keywords'].str.lower()

# Function to categorize a transaction based on similarity
def categorize_transaction(transaction_description):
    transaction_description = transaction_description.lower()
    for index, row in categories_df.iterrows():
        if any(keyword.lower() in transaction_description for keyword in row['Keywords'].split(',')):
            return row['Category']
    return "UNKNOWN"

# Load transactions
transactions_df = pd.read_excel('/content/Book1.xlsx')

# Convert transaction descriptions to lowercase
transactions_df['Description'] = transactions_df['Description'].str.lower()

# Categorize transactions
transactions_df['Category'] = transactions_df['Description'].apply(categorize_transaction)

# Handle unknown categories
unknown_transactions = transactions_df[transactions_df['Category'] == 'UNKNOWN']
for index, row in unknown_transactions.iterrows():
    print(f"Transaction: {row['Description']}")
    category = input("Enter category: ")
    transactions_df.loc[index, 'Category'] = category

    # Update categories and keywords
    keyword = input("Enter keyword from the transaction for future reference: ")
    new_category_row = pd.DataFrame([[category, keyword.lower()]], columns=['Category', 'Keywords'])
    categories_df = pd.concat([categories_df, new_category_row], ignore_index=True)
    categories_df.to_csv('/content/categories_and_keywords.csv', index=False)

# Save categorized transactions to a new Excel file
transactions_df.to_excel('/content/CategorizedTransactions.xlsx', index=False)

# Download the updated Excel file
files.download('/content/CategorizedTransactions.xlsx')

# Visualization
def visualize_data():
    plt.figure(figsize=(10, 6))

    # Plot the distribution of categories
    plt.subplot(1, 2, 1)
    sns.countplot(data=transactions_df, x='Category', order=transactions_df['Category'].value_counts().index)
    plt.title('Transaction Categories Distribution')
    plt.xticks(rotation=45)


    plt.tight_layout()
    plt.show()

visualize_data()

# Function to retrain and update the model with new data
def retrain_model():
    # Reload the updated CSV and Excel files
    global categories_df
    categories_df = pd.read_csv('/content/categories_and_keywords.csv')

    # Convert keywords to lowercase
    categories_df['Keywords'] = categories_df['Keywords'].str.lower()

    # Re-categorize transactions with updated keywords
    transactions_df = pd.read_excel('/content/Book1.xlsx')
    transactions_df['Description'] = transactions_df['Description'].str.lower()
    transactions_df['Category'] = transactions_df['Description'].apply(categorize_transaction)

    # Save updated transactions
    transactions_df.to_excel('/content/CategorizedTransactions_Updated.xlsx', index=False)
    files.download('/content/CategorizedTransactions_Updated.xlsx')

retrain_model()
```

```python
import pandas as pd
import spacy
from sklearn.metrics import accuracy_score
from google.colab import files

# Load spaCy model
nlp = spacy.load('en_core_web_sm')

# Load categories and keywords
categories_df = pd.read_csv('/content/categories_and_keywords (1).csv')

# Convert keywords to lowercase for case-insensitive comparison
categories_df['Keywords'] = categories_df['Keywords'].str.lower()

# Function to categorize a transaction based on similarity
def categorize_transaction(transaction_description):
    transaction_description = transaction_description.lower()
    for index, row in categories_df.iterrows():
        if any(keyword.lower() in transaction_description for keyword in row['Keywords'].split(',')):
            return row['Category']
    return "UNKNOWN"

# Load transactions with true categories for accuracy calculation
# Use the provided updated file
test_transactions_df = pd.read_excel('/content/CategorizedTransactions.xlsx')

# Convert transaction descriptions to lowercase
test_transactions_df['Description'] = test_transactions_df['Description'].str.lower()

# Predict categories
test_transactions_df['Predicted_Category'] = test_transactions_df['Description'].apply(categorize_transaction)

# Calculate accuracy
true_categories = test_transactions_df['Category']
predicted_categories = test_transactions_df['Predicted_Category']

accuracy = accuracy_score(true_categories, predicted_categories) * 100
print(f"Model Accuracy: {accuracy:.2f}%")

# Save results to a new Excel file for review
test_transactions_df.to_excel('/content/AccuracyResults.xlsx', index=False)

# Download the accuracy results file
files.download('/content/AccuracyResults.xlsx')
```
